# Graphes 3D

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris =load_iris()
x = iris.data
y =iris.target
names = list(iris.target_names)
print(f'x contient {x.shape[0]} exemples et {x.shape[1]} variables')
print(f'Il y\'a {np.unique(y).size} classes')
```

```python
y
```

```python
plt.scatter(x[:,0],x[:,1], c=y, alpha=0.5, s=x[:,3]*100)
plt.xlabel('Longueur Sépal')
plt.ylabel('Largeur Sépal')
plt.show()
```

```python
#%matplotlib
from mpl_toolkits.mplot3d import Axes3D
```

```python
G = plt.axes(projection='3d')
G.scatter(x[:,0],x[:,1],x[:,2], c=y)
plt.show()
```

```python
#x =[i for i in range(-10000,10000)]
f = lambda x, y: np.sinc(x+y)
```

```python
x = np.linspace(-40,40,100)
y =np.linspace(-40,40,100)
```

```python
x,y = np.meshgrid(x,y)
```

```python
f(x,y).shape
```

```python
GF = plt.axes(projection='3d')
GF.plot_surface(x,y,f(x,y), cmap='plasma')
plt.show()
```

```python
f =lambda x, y:np.sinc(x) +np.cos(x+y)
x = np.linspace(0,5,100)
y =np.linspace(0,5,100)
x,y = np.meshgrid(x,y)
z =f(x,y)
```

```python
z.shape
```

```python
FG = plt.axes(projection='3d')
FG.plot_surface(x,y,z, cmap='plasma')
plt.show()
```

# Contour plot

```python
plt.contour(x,y,z,20)
plt.show()
```

```python
plt.contourf(x,y,z)
plt.colorbar()
plt.show()
```
