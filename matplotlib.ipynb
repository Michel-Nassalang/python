```python
dataset = {f"experience{i}" : np.random.randint(0,10, [20,20]) for i in range(5)}
dataset
```

```python
def graphique(dataset):
    plt.figure(figsize=(15,30))
    fig, G = plt.subplots(4,1, sharex=True)
    for k, i in zip(dataset.keys(), range(4)):
        G[i].plot(dataset[f'experience{i}'][i])
        G[i].plot(dataset[f'experience{i}'][i+1])
    plt.show()
```

```python
graphique(dataset)
```
