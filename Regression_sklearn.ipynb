```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
np.random.seed(0)
x = np.random.randint(15,125, 200)
```

```python
np.random.seed(0)
f  = lambda x : x + 2*np.random.randn(x.shape[0])
```

```python
y=f(x)
```

```python
y.shape
```

```python
x.shape
```

```python
y = y.reshape(200,1);
x = x.reshape(200,1);
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y, c='g')
plt.show()
```

```python
x = x.reshape(x.shape[0],1)
y = y.reshape(y.shape[0],1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
modele = LinearRegression()
modele.fit(x,y)
modele.score(x,y)
```

```python
L_predictions = modele.predict(x)
```

```python
plt.figure(figsize=(7,5))
plt.scatter(x, y, c='b')
plt.plot(x, L_predictions, c='black', lw=4)
plt.show()
```

```python
from sklearn.svm import SVR
```

```python
model = SVR(C=100)
model.fit(x, y)
model.score(x, y)
```

```python
predictions = model.predict(x)
```

```python
plt.figure(figsize=(7,5))
plt.scatter(x, y, c='g')
plt.plot(x, predictions, c='r')
plt.show()
```
